# Basic Line Chart


```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

##Line 1
x = [1,2,3,4,5,6,7,8,9]
y = [i**3 for i in x]

#resize the graph
#figsize gives the ratio of the graph
#the dpi gives the pixels and it is better to have a dpi of >300
#because lower than 300 will give very pixelated lines and graph
plt.figure(figsize = (6,2), dpi=100)

##Line 2
#here using numpy we are creating a range of number
#select interval we want to plot points at
x1 = np.arange(0,9.5,0.5)
#plotting it on the same graph
plt.plot(x1[:6],x1[:6]**2, 'r.-',label='Square')
plt.plot(x1[5:],x1[5:]**2,'y--')
#using the indexing method you can choose values that need to be 
#selected from the list or the values you want to represent


#Keyword Argument Notation
#plt.plot(x,y, label='Cubes',color='#13a5a3', marker='o',
#         linewidth=4, markersize=8,linestyle='-', markededgecolor = 'blue')

# here we can also use the short hand format
#short hand notation(check plot documentation)
#format = '[color][marker][line]'
plt.plot(x,y, 'b.-',label='Cubes')

plt.title('First Graph',fontdict={'fontname':'Arial',
                                  'fontsize':20})

plt.xlabel('x values in range(1,10)', fontsize=14)
plt.ylabel('Cubes',fontsize=14)

#the number displaying the values are called tickmarks

#using the below argument xticks and then ticks, we can create a list
# of all the values you want to see in the xaxis, also using the fontsize
# we can change the font size, 
plt.xticks(ticks=[0,5,9],fontsize = 15)

#how to place a legend
plt.legend()

#SAVE A GRAPH
plt.savefig('mygraph.png', dpi=300)

plt.show()
```

# Pie Chart

```python
#count number of people playing with their left foot 
#using the .loc to lookup by filtering the data set based 
#on a condition
left = fifa.loc[fifa['Preferred Foot']== 'Left'].count()[0]
#the code on top first uses the .loc to locate the condition and then
#it checks it with the condition and then it counts the number of true
# now to just get the count using count()[0] ensures the first row 
#has been called which will only give us a count
#this commamd return a dataframe and we grab the first row to get the count
right = fifa.loc[fifa['Preferred Foot']== 'Right'].count()[0]
```

```python
right
```

```python
left
```

```python
#how to plot a pie
#always pass in number to be easier in plotting
#always pass in a list for a better outcome avoiding error

#labels that give us a better idea of how to present the different variations 
labels = ['Left','Right']

colors = ['red','yellow']
#always have your labels ordered to ensure minimal problem while labelling

#autopct is autopercentage,autopct= syntax for percentage 
#with '' because its a string,then formula = %.2f%% where 2f is 2float points
plt.pie([left,right],labels=labels,colors = colors, autopct='%.2f%%')

#labeling of chart

plt.title('Preferred Foot')

plt.show()
```

# Pie Chart-2


```python
fifa.Weight
type(fifa.Weight)
'''so here we saw the different weight categories for the weight and now
we first need to remove the lbs from the string. Here as Keith said we 
can use strip method to remove the "lbs " '''

'''So here we changed the fifa.Weight dataframe by using strip to remove the lbs 
then we saw an error because of missing or different values, so we added an
if statement to only use the strip on str values, and to get an integer output we
converted it to int'''
fifa.Weight = [int(x.strip('lbs')) if type(x)==str else x for x in fifa.Weight]
```

```python
fifa.Weight

#Now here to make the pie chart viable we nee to make different categories to
# make the representation more presentable.

#here like earlier we can use the fifa.loc to locate the the values and filter
# it using a condition, if the condition is fulfilled then it get stored in the
# variable which can then be further used to form a pie chart
light = fifa.loc[fifa.Weight < 125].count()[0]
#here we use parenthesis so that we can use logical operators 
light_medium = fifa.loc[(fifa.Weight>=125) & (fifa.Weight < 160)].count()[0]
medium = fifa.loc[(fifa.Weight>=160) & (fifa.Weight < 185)].count()[0]
medium_heavy = fifa.loc[(fifa.Weight>=185) & (fifa.Weight < 200)].count()[0]
heavy = fifa.loc[fifa.Weight>=200].count()[0]
```

```python
#incorporating into a pie chart
labels = ['<125','125-160','160-185','185-200','>200']

#how to change the color scheme using the style.use

plt.style.use('ggplot')

#here we can use pctdistance to set the percentage at a particular distance 
# form the center of the pie chart
#also we can use explode to separate the graph from the center to have a better 

#visualization of the different pies,creat a tuple for explode
#we need to only explode those pies which are difficult to differentiate and 
#here the <125 and >200 
explode = (0.8,0,0,0.1,0.5)

#title
plt.title('Weight Distribution of FIFA Players(in Lbs)')
#plotting the pie
plt.pie([light,light_medium,medium,medium_heavy,heavy],labels=labels,
        autopct='%.2f%%',explode = explode)

plt.show()
```

# Box and Whisker

```python
'''Here we are going to find out and compare the overall scores of the team
players of various teams
firstly we are going to use the .loc command to locate and filter out the 
resp. aspect of the player and multiple teams'''

barca = fifa.loc[fifa.Club == 'FC Barcelona']['Overall']
juve = fifa.loc[fifa.Club == 'Juventus']['Overall']
real = fifa.loc[fifa.Club == 'Real Madrid']['Overall']
psg = fifa.loc[fifa.Club == 'Paris Saint-Germain']['Overall']  
man_u = fifa.loc[fifa.Club == 'Manchester United']['Overall'] 
man_c = fifa.loc[fifa.Club == 'Manchester City']['Overall']
chelsea = fifa.loc[fifa.Club == 'Chelsea']['Overall']

teams= [barca,juve,real,psg,man_u,man_c,chelsea]
labels = ['FC Barcelona','Juventus','Real Madrid','Paris Saint-Germain'
         ,'Manchester United', 'Manchester City','Chelsea']
```

```python
#here we are plotting a box and whiskers using a plt.boxplot

plt.figure(figsize = (14,9))
#to add changes to color and look of the box and whiskers use the 
#patch_artist = True
boxes = plt.boxplot(teams,labels = labels,patch_artist = True)

#syntax to change the color scheme
for box in boxes['boxes']:
    #set edge color
    box.set(color = '#4286f4',linewidth = 3)
    
    #change fill color
    box.set(facecolor = '#e0e0e0')
plt.title('Professional Soccer Team Comparison',fontdict={'fontsize': 20})
plt.ylabel('FIFA Overall Rating')

plt.show()
```
